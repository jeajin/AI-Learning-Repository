```python
import numpy as np 
import matplotlib.pylab as plt 
```

```python
def sigmoid(x): #graph of sigmoid
        return 1/(1+np.exp(-x)*1)
    
def step_function(x): #graph of step function
    return np.array(x>0,dtype=np.int)    

def relu(x): #graph of step function
    return np.maximum(0,x)

x = np.arange(-5.0,5.0,0.1)
y1 = step_function(x)
y2 = sigmoid(x)
y3 = relu(x)
plt.plot(x,y1,label="step")
plt.plot(x,y2,linestyle="--",label="sigmoid")
plt.plot(x,y3,linestyle=":",label="relu")
plt.ylim(-0.1,1.1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("step & sigmoid & relu")
plt.legend()
plt.show()
```
